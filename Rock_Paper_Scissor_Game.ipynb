```python
import random
from PIL import Image
import requests
from io import BytesIO
from IPython.display import display  # For displaying

def fetch_and_display_image(url):
    response = requests.get(url)
    img = Image.open(BytesIO(response.content))
    return img.resize((100, 100))

def display_images(game_img, rock_img, paper_img, scissor_img):
    display(game_img)
    display(rock_img)
    display(paper_img)
    display(scissor_img)

def menu():
    print("\t1. Play Rock, Paper, Scissors with computer")
    print("\t2. Exit game")

def rps():
    game_url = "https://www.shutterstock.com/image-vector/rock-paper-scissors-icon-set-260nw-1804156864.jpg"
    rock_url = "https://rock-paper-scissor-game-web-game.netlify.app/image/stones.png"
    paper_url = "https://www.vhv.rs/dpng/d/535-5351099_rock-paper-scissors-rock-paper-scissors-icons-png.png"
    scissor_url = "https://rock-paper-scissor-game-web-game.netlify.app/image/scissors2.png"

    game_img = fetch_and_display_image(game_url)
    rock_img = fetch_and_display_image(rock_url)
    paper_img = fetch_and_display_image(paper_url)
    scissor_img = fetch_and_display_image(scissor_url)

    display_images(game_img, rock_img, paper_img, scissor_img)

    while True:
        menu()
        choice = input("Enter your choice: ")
        if choice == '1':
            play_game(rock_img, paper_img, scissor_img)
        elif choice == '2':
            print("Exiting game...")
            break
        else:
            print("Invalid choice. Please enter 1 or 2.")

def play_game(rock_img, paper_img, scissor_img):
    choices = {'r': 'rock', 'p': 'paper', 's': 'scissor'}
    image_mapping = {'r': rock_img, 'p': paper_img, 's': scissor_img} # Map choices to image variables
    p1_choice = input("Enter your choice (r/rock, p/paper, s/scissor): ").lower()
    if p1_choice not in choices:
        print("Invalid choice.")
        return

    p1_name = choices[p1_choice]
    print(f"Your choice: {p1_name}")
    display(image_mapping[p1_choice]) # Use the mapping to display the correct image

    com_choice = random.choice(list(choices.keys()))
    com_name = choices[com_choice]
    print(f"Computer's choice: {com_name}")
    display(image_mapping[com_choice]) # Use the mapping for computer's choice as well

    if p1_choice == com_choice:
        print("It's a tie!")
    elif (p1_choice == 'r' and com_choice == 's') or \
         (p1_choice == 'p' and com_choice == 'r') or \
         (p1_choice == 's' and com_choice == 'p'):
        print("You win!")
    else:
        print("Computer wins!")

# Start the game
rps()
```
