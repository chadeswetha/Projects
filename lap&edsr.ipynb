EDSR

```python
import cv2
from cv2 import dnn_superres
```

```python
import cv2
import matplotlib.pyplot as plt

# Initialize super resolution object
sr = cv2.dnn_superres.DnnSuperResImpl_create()

# Read the model
path = '/content/EDSR_x4 (5).pb'
sr.readModel(path)

# Set the model and scale
sr.setModel('edsr', 4)

# If you have CUDA support
sr.setPreferableBackend(cv2.dnn.DNN_BACKEND_CUDA)
sr.setPreferableTarget(cv2.dnn.DNN_TARGET_CUDA)

# Load the image
image = cv2.imread('/content/test2134 (1).png')

# Upsample the image
upscaled = sr.upsample(image)

# Save the upscaled image
#cv2.imwrite('upscaled_test1ESDR.png', upscaled)

# Display the input and output images
plt.figure(figsize=(10, 5))

# Input image
plt.subplot(1, 2, 1)
plt.title('Input Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Output image
plt.subplot(1, 2, 2)
plt.title('EDSR Image')
plt.imshow(cv2.cvtColor(upscaled, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()

```

LapSRN

```python
import cv2
import matplotlib.pyplot as plt

# Initialize super resolution object
sr = cv2.dnn_superres.DnnSuperResImpl_create()

# Read the model
path = '/content/LapSRN_x8.pb'
sr.readModel(path)

# Set the model and scale - match the model name and scale
sr.setModel('lapsrn', 8)  # Change to 'lapsrn' and scale 8

# If you have CUDA support
sr.setPreferableBackend(cv2.dnn.DNN_BACKEND_CUDA)
sr.setPreferableTarget(cv2.dnn.DNN_TARGET_CUDA)

# Load the image
image = cv2.imread('/content/test2134 (1).png')

# Upsample the image
upscaled = sr.upsample(image)

# Save the upscaled image
#cv2.imwrite('upscaled_test1ESDR.png', upscaled)

# Display the input and output images
plt.figure(figsize=(10, 5))

# Input image
plt.subplot(1, 2, 1)
plt.title('Input Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Output image
plt.subplot(1, 2, 2)
plt.title('LapSRN Image') # Update title
plt.imshow(cv2.cvtColor(upscaled, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()
```
